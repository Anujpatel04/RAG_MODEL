```python
import requests

url = "https://dspmuranchi.ac.in/pdf/Blog/Python%20Built-In%20Functions.pdf"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}
response = requests.get(url, headers=headers)

if response.status_code == 200:
    with open("python_functions.pdf", "wb") as f:
        f.write(response.content)
    print("PDF downloaded successfully")
else:
    print(f"Error: {response.status_code}, {response.reason}")
```

```python
from langchain.document_loaders import PyPDFLoader

pdf_path = "python_functions.pdf"
loader = PyPDFLoader(pdf_path)
documents = loader.load()

print(f"Loaded {len(documents)} pages successfully!")
```

```python
text_splitter = CharacterTextSplitter(
    chunk_size=2000,
    chunk_overlap=400
)
```

```python
texts = text_splitter.split_documents(documents)
```

```python
embeddings = HuggingFaceEmbeddings()
```

```python
persist_directory = "vector_db"
```

```python
vectordb = Chroma.from_documents(
    documents=texts,
    embedding=embeddings,
    persist_directory=persist_directory
)
```

```python
# retriever
retriever = vectordb.as_retriever()
```

```python
# llm from groq
llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0
)
```

```python
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=retriever,
    return_source_documents=True
)
```

```python
# invoke the qa chain and get a response for user query
query = "what are the function from this pdf "
response = qa_chain.invoke({"query": query})
```

```python
print(response["result"])
```

```python
print(response["source_documents"][0].metadata["source"])
```
